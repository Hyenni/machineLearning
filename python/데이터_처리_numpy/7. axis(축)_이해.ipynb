### 학습목표
#### axis의 이해
#### axis를 파라미터로 갖는 함수 이해하기

```python
import numpy as np
```

#### axis 이해하기
 - 몇몇 함수에는 axis keyword 파라미터가 존재
 - axis값이 없는 경우에는 전체 데이터에 대해 적용
 - axis값이 있는 경우에는, 해당 axis를 **따라서** 연산 적용

* axis를 파라미터로 갖는 함수를 이용하기
 - 거의 대부분의 연산 함수들이 axis 파라미터를 사용
 - 이 경우, 해당 값이 주어졌을 때, 해당 axis를 **따라서** 연산이 적용
   - 따라서 결과는 해당 axis가 제외된 나머지 차원의 데이터만 남게 됨
 - 예) np.sum, np.mean, np.any 등등
 

```python
x = np.arange(15)
print(x)
```

* 1차원 데이터에 적용하기

```python
np.sum(x, axis=0)
```

1차원 데이터이기 때문에 axis가 0 밖에 없음. 따라서 1로 설정했을 때 에러 발생

```python
np.sum(x, axis=1)
```

* 행렬에 적용하기

```python
y = x.reshape(3, 5)
print(y)
```

```python
np.sum(y, axis=0)
```

```python
np.sum(y, axis=1)
```

* 3차원 텐서에 적용하기

3차원은 순서대로 axis가 차원, 행, 열을 의미한다고 생각하면 된다.

```python
z = np.arange(36).reshape(3, 4, 3)
print(z)
```

axis = 0은 차원을 따라서 압축하는 것. 

또한 기존 (3, 4, 3)의 shape에서 0번째 축을 따라서 압축했으므로 남는 것은 (4, 3)의 2차원 행렬이 된다.


![axis0](https://user-images.githubusercontent.com/33708512/135740361-cdac33f8-ecd9-41c3-bb5c-fce3e63936e5.png)

```python
np.sum(z, axis=0)
```

axis = 1은 행을 따라 압축하는 것.


![axis1](https://user-images.githubusercontent.com/33708512/135740362-ff0b178e-a0b2-44e0-894b-e4004eba17ad.png)


```python
np.sum(z, axis=1)
```

axis = 2는 열을 따라 압축하는 것.



```python
np.sum(z, axis=2)
```

```python
np.sum(z, axis=-3)
```

* axis의 값이 튜플일 경우
 - 해당 튜플에 명시된 모든 axis에 대해서 연산
 
 이처럼 튜플로 있을 때, 순서대로 하면 압축하면 됨.
 
 3차원에서 차원을 따라 합한뒤 2차원 데이터에서 행을 따라 합하면 되는 것이다.

 

```python
print(z)
```

```python
np.sum(z, axis=(0, 2))
```
