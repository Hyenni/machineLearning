### 학습목표
#### DataFrame groupby 이해하기

```python
import pandas as pd
import numpy as np
```

#### group by
  + 아래의 세 단계를 적용하여 데이터를 그룹화(groupping) (SQL의 group by 와 개념적으로는 동일, 사용법은 유사)
    - 데이터 분할
    - operation 적용
    - 데이터 병합

```python
# data 출처: https://www.kaggle.com/hesh97/titanicdataset-traincsv/data
df = pd.read_csv('./train.csv')
df.head()
```

#### GroupBy groups 속성
 - 각 그룹과 그룹에 속한 index를 dict 형태로 표현

```python
class_group = df.groupby('Pclass')
class_group
```

```python
class_group.groups
```

```python
gender_group = df.groupby('Sex')
gender_group.groups
```

#### groupping 함수
 - 그룹 데이터에 적용 가능한 통계 함수(NaN은 제외하여 연산)
 - count - 데이터 개수 
 - sum   - 데이터의 합
 - mean, std, var - 평균, 표준편차, 분산
 - min, max - 최소, 최대값

```python
class_group.count()
```

```python
class_group.mean()
```

```python
class_group.mean()['Age']
```

```python
class_group.mean()['Survived']   #생존률
```

* 성별에 따른 생존율 구해보기

```python
df.groupby('Sex').mean()
```

```python
df.groupby('Sex').mean()['Survived']
```

#### 복수 columns로 groupping 하기
 - groupby에 column 리스트를 전달
 - 통계함수를 적용한 결과는 multiindex를 갖는 dataframe

* 클래스와 성별에 따른 생존률 구해보기

```python
class_sex = df.groupby(['Pclass', 'Sex'])
class_sex
```

```python
class_sex.groups
```

```python
class_sex.mean()
```

```python
class_sex.mean()['Survived']
```

```python
class_sex.mean().index
```

index가 두 개이기 때문에 loc을 이용해서 행을 가져오기 위해서는 두 개 모두 명시해줘야 함.

```python
class_sex.mean().loc[(2, 'female')]
```

#### index를 이용한 group by
 - index가 있는 경우, groupby 함수에 level 사용 가능
   - level은 index의 depth를 의미하며, 가장 왼쪽부터 0부터 증가

* **set_index** 함수
 - column 데이터를 index 레벨로 변경
* **reset_index** 함수
 - 인덱스 초기화

기본적으로 0-based index임.

```python
df.head()
```

```python
df.set_index(['Pclass'])
```

```python
df.set_index(['Pclass', 'Sex'])
```

```python
df.set_index(['Pclass', 'Sex']).reset_index()
```

index가 있는 경우에 groupby에 해당 column명을 쓰지 않고 level을 이용할 수 있음. 

```python
df.set_index('Age').groupby(level=0).mean()      


## df.groupby('Age').mean() 인덱스 없는 경우
```

#### 나이대별로 생존율 구하기

```python
import math
def age_categorize(age):
    if math.isnan(age):
        return -1
    return math.floor(age / 10) * 10
```

index가 있는 경우에 한해서 groupby 안에 함수를 사용할 수 있음.

age_categorize 함수에 의해 계산된 값을 기준으로 그룹핑됨.

```python
df.set_index('Age').groupby(age_categorize)
```

```python
df.set_index('Age').groupby(age_categorize).mean()
```

```python
df.set_index('Age').groupby(age_categorize).mean()['Survived']
```

#### MultiIndex를 이용한 groupping

```python
df.set_index(['Pclass', 'Sex']).groupby(level=[0, 1]).mean()['Age']
```

#### aggregate(집계) 함수 사용하기
 - groupby 결과에 집계함수를 적용하여 그룹별 데이터 확인 가능
 
 한 개의 통계적 자료만 궁금하면 바로 호출하면 되는데 (.mean() 등)
 리스트 형태로 한 번에 여러 개의 통계적 자료가 궁금하면 aggregate를 사용하면 됨.
 

```python
df.set_index(['Pclass', 'Sex']).groupby(level=[0, 1]).aggregate([np.mean, np.sum, np.max])
```
